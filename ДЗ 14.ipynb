# Задание 3

```python
from scipy.linalg import solve
import matplotlib.pyplot as plt
plt.style.use('ggplot')
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
x = np.linspace(1,15,100)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(18,12), linewidth=2)
ax.plot(x, f(x), label='Сама функция')
for points in map(np.array,([1,15],[1,8,15],[1,4,10,15])):
  weights = solve(np.vander(points), f(points))
  ax.plot(x, np.polyval(weights,x), label = 'Аппроксимация {}-го порядка, веса: {}'.format(weights.shape[0], weights), linewidth=3, linestyle='--')
ax.legend();
```
